### Matplotlib : Tracer une courbe et effectuer une regression lin√©aire
</div>

```python
# Chargement des biblioth√®ques scientifiques
import matplotlib.pyplot as plt
import numpy as np

plt.clf()

# Quelques valeurs
X = np.array([0, 0.021, 0.039, 0.063])
Y = np.array([0, 0.84, 1.45, 2.4])

# On trace Y= f(X)
plt.plot(X, Y, 'x', color='blue')  


plt.xlabel("L√©gende des abscisses")
plt.ylabel("L√©gende des ordonn√©es")
plt.title("Titre du graphique")
plt.grid()

# Demande √† Numpy de trouver les coefficients a et b pour Y = a X + b
fit = np.polyfit(X, Y, 1)

coefficient_directeur = fit[0] # Coefficient directeur - a 
ordonnee_origine = fit[1] # Ordonn√©e origine - b
print("Ordonn√©e √† l'origine:", ordonnee_origine)
print("Coefficient directeur :", coefficient_directeur)
print("Equation : Y = " + str(coefficient_directeur) + ' X + ' + str(ordonnee_origine) )
correlation = np.corrcoef(X, Y)[0,1]
r_carre = correlation**2
print("r_carre (proche de 1 si mod√©lisation de bonne qualit√©) :",r_carre)

# On calcule la position des Y en fonction des X suivant l'√©quation mod√©lis√©e Y = a X + b
Modelisation = coefficient_directeur * X + ordonnee_origine
# On trace
plt.plot(X, Modelisation, color='red')
# On affiche
plt.show()
```

En Python, <code>Si condition alors</code> s'√©crira <code>if condition:</code> suivi d'un bloc indent√©. <code>Sinon</code> s'√©crit simplement <code>else:</code>.<br>
<b>Attention:</b> Pour tester l'√©galit√© en Python, il faut √©crire par exemple <code>if x==2:</code>: On met <b>deux</b> fois le symbole =.<br>
Voici le m√™me algorithme retranscrit en langage Python:

```python
def f(x):
    if x>=0:    #  >= signifie sup√©rieur ou √©gal
        y=x     #  Remarquons que cette instruction est indent√©e pour que Python reconnaisse ce qui doit √™tre execut√© dans le if
    else:
        y=-x
    return(y)   #  Cette instruction n'est pas indent√©e comme la ligne pr√©c√©dente: elle est execut√©e apr√®s les tests.

# Si x est positif, on le retourne, sinon on retourne son oppos√©. Ce programme supprime le - d'un nombre
```

<h2 style="text-decoration:underline;" id="partie3">3- Notion d'accumulateur</h2>

Le compteur de boucle permet de mettre en place un accumulateur, c'est-√†-dire une variable qui est utilis√©e pour stocker les diff√©rents r√©sultats successifs d'un calcul qui avance progressivement √† chaque tour de boucle. On √©conomise ainsi de la m√©moire : on ne garde en m√©moire par exemple que la somme des valeurs d√©j√† rencontr√©es ou le dernier √©tat, et non tous les √©tats interm√©diaires.<br>
La variable qui sert d'accumulateur doit √™tre initialis√©e avant la boucle.

<strong>Exemple : calcul de la somme des entiers de 1 √† 100</strong>

```python
somme = 0 # La variable somme joue le r√¥le de l'accumulateur
for i in range(101):
    somme = somme + i   # √† chaque tour de boucle, on met √† jour la valeur de l'accumulateur en lui ajoutant i
print(somme)      # On affiche la valeur finale de l'accumulateur
```

Remarque : il existe une formule qui permet de calculer cette somme sans effectuer toutes les additions, qui sera vue en cours de Sp√©cialit√© Maths : $1+2+3+...+n = \frac{n(n+1)}{2}$.

**Fonctions sans valeur de retour**<br>
Une fonction peut ne renvoyer aucune valeur. C'est le cas par exemple de la fonction <code>print()</code> ou des fonctions de trac√© du module <ocde>turtle</code>. Le r√¥le d'une telle fonction est d'effectuer une travail qui ne n√©cessite pas de renvoyer une information.<br>

Dans le cas d'une fonction sans valeur de retour, on peut omettre l'instruction <code>return</code>.

```python
def disCoucou(n):
    for i in range(n):
        print("Coucou!")
    return     # l'instruction return est facultative ici

disCoucou(3)
print()  # affiche une ligne vide pour s√©parer les affichages
a = disCoucou(2)
print(a)
```

<div class="alert alert-warning" style="border-left: 3px solid"><strong>Exercice 58 : </strong> la fonction <code>triangle()</code> ci-dessous permer de dessiner un triangle dont la base est horizontale et dont les coordonn√©es du sommet sont (x,y).<br>
    1) En utilisant la fonction <code>triangle()</code>, cr√©er une fonction <code>sapin(x,y)</code> qui dessine un sapin tel que celui dessin√© ci-dessous.<br>
    2) Cr√©er une fonction <code>for√™t(n)</code> qui dessine une for√™t de <code>n</code> sapins dont la position est choisie al√©atoirement.

```python
from turtle import *
from random import *

def triangle(x,y,c√¥t√©,couleur):
    penup()
    goto(x,y)
    pendown()
    begin_fill()
    color('black',couleur)
    left(210)
    forward(c√¥t√©)
    left(150)
    forward(c√¥t√©*1.732)
    left(150)
    forward(c√¥t√©) 
    left(210)
    end_fill()
```

```python
def sapin(x,y):
    for i in range(5):
        triangle(x, y - i * 15, 50,'brown')
    for i in range(10):
        triangle(x, y + 225 - i * 20, 150,'green')

speed(10)
sapin(0,0)
done()
```

```python
def for√™t(n):
    for i in range(n):
        sapin(-200+random()*400,(n-i)/n*100-100)
        
speed(0)
for√™t(20)
done()
```

**R√©cursivit√©** *<br>
Une fonction peut faire appel √† elle-m√™me dans sa propre d√©finition. On parle alors de fonction <strong>r√©cursive</strong>. La r√©cursivit√© sera √©tudi√©e en Terminale Sp√©cialit√© NSI.<br>
La fonction suivante calcule la somme des entiers de 0 √† n de fa√ßon r√©cursive :

```python
def somme(n):
    if n==0:
        return 0
    else:
        return n+somme(n-1)

somme(100)
```

**Port√©e des variables**

Les variables qui sont cr√©√©es dans la d√©claration d'une fonction n'existent (ne sont accessibles) que dans le corps de la fonction : ce sont des **variables locales**.<br>
Ce ph√©nom√®ne peut √™tre v√©rifi√© sur l'exemple suivant :

```python
def maFonction(a):
    b = 3
    r√©sultat = a + b
    return r√©sultat

print(maFonction(5))
print(b)     # erreur : la variable b n'existe pas en dehors du corps de la fonction maFonction()
```

Voici une variante de l'exemple pr√©c√©dent :

```python
def maFonction(a):
    b = 3
    r√©sultat = a + b
    print(b)     # il s'agit ici de la variable locale b, qui vaut ici 3
    return r√©sultat

b = 10
print(maFonction(5))
print(b)     # il s'agit ici de la variable globale b, qui vaut 10
             # Cette variable b n'a pas √©t√© modifi√©e lors de l'appel mafonction(5) 
```

Si l'on souhaite utiliser une variable globale dans le corps d'une fonction, on doit le pr√©ciser avec le mot-cl√© <code>global</code>. Sinon, la variable sera consid√©r√©e comme locale.<br>
Voici un exemple :

```python
def maFonction(a):
    global b # on indique ici que b est une variable globale dans tout le corps de cette fonction
             # b d√©signe donc la variable qui a √©t√© d√©finie plus haut et dont la valeur est 10
    print(b)
    b = 3      # la valeur de la variable globale b est √©cras√©e et devient 3
    r√©sultat = a + b
    print(b)     
    return r√©sultat

b = 10    # Variable globale, d√©clar√©e en dehors du corps de la fonction maFonction
print(b)
print(maFonction(5))
print(b)     # on constate que la valeur de b a √©t√© modifi√©e apr√®s l'appel maFonction(5)
```

Les param√®tres formels d'une fonction se comportent eux aussi comme des variables locales, comme le montre l'exemple suivant :<br>

```python
def carr√©(x):
    x = 4      # on peut modifier la valeur de x dans le corps de la fonction, x est une variable locale
    print(x)
    return x*x   # par cons√©quent, la fonction ne renvoie pas le carr√© du param√®tre x car la valeur de x a √©t√© modifi√©e

print(carr√©(10))
```

**Annotations de type** *
<p style="text-align:justify;">Dans de nombreux langages (comme Java ou C++), le type des param√®tres et de la valeur de retour d'une fonction doivent √™tre pr√©cis√©s dans la d√©claration de la fonction. Cette contrainte permet une d√©tection plus pr√©coce de certaines erreurs, avant m√™me d'ex√©cuter la fonction.<br>
Les derni√®res versions de Python donnent la possibilit√© d'ajouter des annotations de type dans la premi√®re ligne de la d√©claration d'un fonction.<br>
Dans l'exemple ci-dessous, on indique que le param√®tre <code>n</code> devrait √™tre de type <code>int</code> et que la fonction devrait renvoyer une valeur de type <code>bool</code>.<br>
Comme on le constate ici, l'appel <code>pair(6.2)</code> ne provoque pas d'erreur ici, mais l'erreur pourrait √™tre d√©tect√©e si on utilise un environnement de programmation autre que Capytale.<br>
M√™me dans le cas o√π les annotations de type ne sont pas prises en compte par l'interpr√©teur Python, ces derni√®res apportent des informations sur la mani√®re d'utiliser la fonction ainsi que sur son r√¥le.</p>

```python
def pair(n: int) -> bool:
    if n%2 == 0:
        return True
    else:    
        return False

print(pair(5))
print(pair(6))
print(pair(6.2))
```

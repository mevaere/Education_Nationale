**Introduction**<br>
<p style="text-align:justify;">DRY (Don't Repeat Yourself) est un principe de programmation qui dit que chaque information (ou morceau de code) doit avoir une représentation ou un emplacement unique et ne doit pas être dupliquée.
Ainsi, idéalement, on ne devrait jamais écrire deux fois le même morceau de code.

Les <strong>fonctions</strong> aident le programmeur à respecter ce principe. Ce sont des blocs de code autonomes et réutilisables à volonté. On peut les paramétrer pour leur donner plus de flexibilité.</p>

Voici quelques avantages apportés par les fonctions en programmation :
<ul>
    <li style="text-align:justify;">
        On écrit le code de la fonction une seule fois, puis on peut l'utiliser (l'appeler) autant de fois qu'on le souhaite.
    </li>
    <li style="text-align:justify;">
        La maintenance du programme est facilitée : si plus tard on doit apporter une modification, celle-ci n'est à faire qu'à un seul endroit.   
    </li>
     <li style="text-align:justify;">
        Les fonctions permettent de segmenter le programme, de le découper en petites parties autonomes. Ainsi, la réflexion est facilitée (il est plus facile de se concentrer sur un problème de taille modeste), les tests sont facilités et les risques d'erreurs diminuent.
    </li>
    <li style="text-align:justify;">
        Une conséquence du point précédent est que les fonctions facilitent le travail en équipe en permettant de partager le travail. Chaque personne de l'équipe peut être chargée de s'occuper d'une fonction bien précise et indépendante du reste du programme.
    </li>   
    <li style="text-align:justify;">
        Les fonctions permettent d'étendre le langage en lui ajoutant de nouvelles fonctionnalités. On peut ainsi créer des bibliothèques entières de nouvelles fonctions, que l'on appelle des modules.
    </li>   
</ul>

<h2 style="text-decoration:underline;" id="partie2">2- Créer ses propres fonctions</h2>

Dans tous les langages de programmation, on a la possibilité de créer soi-même ses propres fonctions. La création de fonctions est même l'activité principale des programmeurs.

La syntaxe générale de la déclaration/création d'une fonction en Python est la suivante :<br>
<p style="border:solid;">
<code>def nomDeLaFonction(nomDuParamètre1,nomDuparamètre2,...):
    instructions
    instructions
    return valeur
</code>
</p>
<ul>
<li>La définition d'une fonction en Python commence avec le mot-clé <code>def</code>.</li>
<li>On indique ensuite le nom de nouvelle fonction que l'on souhaite créer.</li>
<li>Les parenthèses qui suivent sont obligatoires : elles indiquent qu'il s'agit d'une fonction et non d'une variable ordinaire.</li>
<li>A l'intérieur des parenthèses, on indique la liste des noms des paramètres de la fonction. Les paramètres sont optionnels : on peut créer une fonction sans paramètres (avec des parenthèses vides). </li>
<li>La première ligne de la définition se termine par le symbole "deux points" (:)</li>
<li>Les instructions qui constituent le coeur de la fonction sont indentées vers la droite. On peut utiliser tout type d'instructions : affectations, boucles, instructions conditionnelles, appels de fonctions.</li>
<li>Le mot-clé <code>return</code> permet de mettre fin à la fonction et/ou de renvoyer une valeur de retour.</li>
<li>Contrairement à d'autres langages, Python n'impose pas d'indiquer à l'avance le type  des paramètres ni celui de la valeur de retour.</li>
</ul>

Voici un exemple de définition de la fonction carré (la même qu'en Mathématiques) : celle-ci renvoie le carré du nombre x donné en paramètre.

```python
def carré(x):
    résultat = x * x
    return résultat
```

On peut aussi écrire de façon plus courte :

```python
def carré(x):
    return x*x
```

<p style="text-align:justify;">Lorsqu'on exécute l'un ou l'autre de ces deux programmes, il ne se passe rien. C'est normal, car on n'a pas encore <strong>appelé</strong> la fonction (ou fait appel à la fonction). On s'est contenté pour le moment de donner sa définition, c'est-à-dire de préciser ce qu'elle doit faire lorsqu'on l'appelle. <br>
Une définition de fonction n'est pas une instruction.<br>
Voici différents appels de cette fonction carré :</p>

```python
carré(7)
```

```python
a = carré(5)     # on stocke la valeur de retour de la fonction carré dans la variable a
print(a)
print(carré(10))   # on affiche directement la valeur de retour de la fonction carré
```

<h2 style="text-decoration:underline;" id="partie3">3- Informations complémentaires sur les fonctions</h2>

<em>Les parties marquées d'une étoile * sont plus avancées et s'adressent aux élèves qui sont déjà à l'aise avec les fonctions en Python. Elles peuvent être omises pour une première découverte des fonctions.</em>

**Fonctions sans valeur de retour**<br>
Une fonction peut ne renvoyer aucune valeur. C'est le cas par exemple de la fonction <code>print()</code> ou des fonctions de tracé du module <ocde>turtle</code>. Le rôle d'une telle fonction est d'effectuer une travail qui ne nécessite pas de renvoyer une information.<br>
Dans l'exemple suivant, la fonction <code>disCoucou()</code> est appelée deux fois. Lors du deuxième appel, la valeur de retour est stockée dans la variable <code>a</code> et on constate que cette valeur est <code>None</code>, ce qui siginife "aucune valeur".<br>
Dans le cas d'une fonction sans valeur de retour, on peut omettre l'instruction <code>return</code>.

```python
def disCoucou(n):
    for i in range(n):
        print("Coucou!")
    return     # l'instruction return est facultative ici

disCoucou(3)
print()  # affiche une ligne vide pour séparer les affichages
a = disCoucou(2)
print(a)
```

**Fonctions sans paramètre**<br>
Une fonction peut n'avoir aucun paramètre, comme par exemple les fonctions <code>print()</code>, <code>input()</code> ou <code>penup()</code> du module <code>turtle</code>. Dans ce cas, on doit tout de même utiliser des parenthèses vides pour indiquer qu'il s'agit bien d'une fonction et non d'une variable ordinaire.<br>
Voici un exemple de fonction sans paramètre :

```python
def pi():     # Cette fonction renvoie la valeur du nombre pi
    return 3.141592653589793
```

**Fonctions à plusieurs paramètres**<br>
Une fonction peut avoir plusieurs paramètres.
Lors de l'appel de la fonction, il faut bien faire attention à respecter le nombre attendu de paramètres, ainsi que l'ordre de ces paramètres.

**Arguments par défaut** *<br>
Dans la définition d'une fonction, il est possible de donner des valeurs par défaut à certains paramètres. Si dans un appel de la fonction, on souhaite utiliser la valeur des paramètres par défaut, il n'st plus nécessaire de préciser la valeur de ces paramètres. 

Dans l'exemple ci-dessous, la fonction <code>écrire()</code> prend un paramètre : une chaîne de caractères <code>message</code> et le nombre de répétitions <code>n</code>. Elle affiche n fois le message.<br>
Le message par défaut est "Bonjour". Le nombre de répétions par défaut est 3.

**Sortie anticipée d'une fonction, cas de plusieurs instructions return**<br>
Il est tout à fait possible d'insérer plusieurs instructions <code>return</code> dans une fonction. Dans ce cas, l'exécution de la fonction s'interrompra dès la première instruction <code>return</code> rencontrée et le programme reviendra à la ligne de code qui a appelé la fonction. Cette possibilité est généralement utilisée conjointement avec des instructions conditionnelles ou des boucles.<br>
L'exemple suivant montre un cas d'utilisation avec la foncion "Valeur absolue" :

```python
def valeurAbsolue(x):
    if x>=0:
        return x
    else:
        return -x  
```

On peut même supprimer le <code>else</code> comme ci-dessous : 

```python
def valeurAbsolue(x):
    if x>=0:
        return x
    return -x
```

**Evaluation paresseuse des opérateurs booléens** *<br>
Les fonctions permettent de vérifier une propriété particulière des opérateurs booléens <code>and</code> et <code>or</code> : <br>
afin d'éviter des calculs inutiles et donc de gagner en performance, une expression booléenne du type <code>A and B</code> va commencer par évaluer la condition <code>A</code>. Dans le cas où cette confition est fausse, il n'est pas nécessaire d'évaluer la condition <code>B</code> car on sait déjà que la valeur de <code>A and B</code> sera <code>False</code>.<br>
Ainsi, la deuxième condition n'est évaluée que si cela est vraiment nécessaire.<br>
Le programme ci-dessous permet de vérifier cette évaluation paresseuse ("lazy evaluation") :

```python
def grand(taille):
    print("Dans la fonction grand()")
    return taille>180

def géant(taille):
    print("Dans la fonction géant()")
    return taille>190

taille=170   # tester aussi avec 185 et 195
print(grand(taille) and géant(taille))       
```

    2) Créer une fonction <code>forêt(n)</code> qui dessine une forêt de <code>n</code> sapins dont la position est choisie aléatoirement.

```python
from turtle import *
from random import *

def triangle(x,y,côté,couleur):
    penup()
    goto(x,y)
    pendown()
    begin_fill()
    color('black',couleur)
    left(210)
    forward(côté)
    left(150)
    forward(côté*1.732)
    left(150)
    forward(côté) 
    left(210)
    end_fill()
```

```python
def sapin(x,y):
    for i in range(5):
        triangle(x, y - i * 15, 50,'brown')
    for i in range(10):
        triangle(x, y + 225 - i * 20, 150,'green')

speed(10)
sapin(0,0)
done()
```

```python
def forêt(n):
    for i in range(n):
        sapin(-200+random()*400,(n-i)/n*100-100)
        
speed(0)
forêt(20)
done()
```

**Récursivité** *<br>
Une fonction peut faire appel à elle-même dans sa propre définition. On parle alors de fonction <strong>récursive</strong>. La récursivité sera étudiée en Terminale Spécialité NSI.<br>
La fonction suivante calcule la somme des entiers de 0 à n de façon récursive :

```python
def somme(n):
    if n==0:
        return 0
    else:
        return n+somme(n-1)

somme(100)
```

**Variables fonctionnelles** *<br>
Une variable peut contenir une fonction, car en Python, une fonction est un objet comme un autre et tout objet peut être stocké dans une variable.<br>
Voici un exemple :

```python
def carré(x):
    return x*x

def cube(x):
    return x*x*x

a=carré          # notez l'absence de parenthèses ici; il ne s'agit pas d'un appel de fonction
print(a(2))
a=cube
print(a(2))
```

Il est même possible que les paramètres d'une fonction (ou la valeur renvoyée par une fonction) soient eux-mêmes des fonctions. Cette pratique s'appelle la **programmation fonctionnelle** et sera évoquée en Terminale Spécialité NSI.

**Documentation de fonction**<br>
<p style="text-align:justify;">On peut documenter une fonction que l'on a créé en lui donnant une chaîne de documentation ("doc string" en anglais).<br>
Il s'agit d'une chaîne de caractères qui explique en quelques mots le rôle de la fonction et éventuellement comment l'utiliser. C'est donc en quelque sorte un mode d'emploi de la fonction. Ecrire une chaîne de documentation n'est pas obligatoire mais c'est une pratique qui est fortement conseillée.<br>
La chaîne de documentation, si elle existe, doit être la première chose déclarée dans la fonction immédiatement après les deux points. Elle doit être écrite entre des triples guillemets, et elle peut occuper sur plusieurs lignes.
Lorsqu'on utilisera plus tard un environnement de développement intégré, la chaîne de documentation apparaitra dans une bulle d'aide lorsqu'on survole le nom de la fonction.<br>
Voici un exemple avec la fonction <code>carré</code> documentée :</p>

```python
def carré(x):
    """Cette fonction renvoie le carré du nombre x.
Elle fonctionne aussi bien avec des int qu'avec des float."""
    return x*x

carré(7)
```

On peut accéder à la doc string depuis un programme de deux manières :

Avec la fonction <code>help()</code> :

```python
help(carré)
```

Avec l'attribut <code>__doc__</code> disponible pour chaque objet :

```python
print(carré.__doc__)
```

**Portée des variables**

Les variables qui sont créées dans la déclaration d'une fonction n'existent (ne sont accessibles) que dans le corps de la fonction : ce sont des **variables locales**.<br>
Ce phénomène peut être vérifié sur l'exemple suivant :

```python
def maFonction(a):
    b = 3
    résultat = a + b
    return résultat

print(maFonction(5))
print(b)     # erreur : la variable b n'existe pas en dehors du corps de la fonction maFonction()
```

Voici une variante de l'exemple précédent :

```python
def maFonction(a):
    b = 3
    résultat = a + b
    print(b)     # il s'agit ici de la variable locale b, qui vaut ici 3
    return résultat

b = 10
print(maFonction(5))
print(b)     # il s'agit ici de la variable globale b, qui vaut 10
             # Cette variable b n'a pas été modifiée lors de l'appel mafonction(5) 
```

Si l'on souhaite utiliser une variable globale dans le corps d'une fonction, on doit le préciser avec le mot-clé <code>global</code>. Sinon, la variable sera considérée comme locale.<br>
Voici un exemple :

```python
def maFonction(a):
    global b # on indique ici que b est une variable globale dans tout le corps de cette fonction
             # b désigne donc la variable qui a été définie plus haut et dont la valeur est 10
    print(b)
    b = 3      # la valeur de la variable globale b est écrasée et devient 3
    résultat = a + b
    print(b)     
    return résultat

b = 10    # Variable globale, déclarée en dehors du corps de la fonction maFonction
print(b)
print(maFonction(5))
print(b)     # on constate que la valeur de b a été modifiée après l'appel maFonction(5)
```

Les paramètres formels d'une fonction se comportent eux aussi comme des variables locales, comme le montre l'exemple suivant :<br>

```python
def carré(x):
    x = 4      # on peut modifier la valeur de x dans le corps de la fonction, x est une variable locale
    print(x)
    return x*x   # par conséquent, la fonction ne renvoie pas le carré du paramètre x car la valeur de x a été modifiée

print(carré(10))
```

**Annotations de type** *
<p style="text-align:justify;">Dans de nombreux langages (comme Java ou C++), le type des paramètres et de la valeur de retour d'une fonction doivent être précisés dans la déclaration de la fonction. Cette contrainte permet une détection plus précoce de certaines erreurs, avant même d'exécuter la fonction.<br>
Les dernières versions de Python donnent la possibilité d'ajouter des annotations de type dans la première ligne de la déclaration d'un fonction.<br>
Dans l'exemple ci-dessous, on indique que le paramètre <code>n</code> devrait être de type <code>int</code> et que la fonction devrait renvoyer une valeur de type <code>bool</code>.<br>
Comme on le constate ici, l'appel <code>pair(6.2)</code> ne provoque pas d'erreur ici, mais l'erreur pourrait être détectée si on utilise un environnement de programmation autre que Capytale.<br>
Même dans le cas où les annotations de type ne sont pas prises en compte par l'interpréteur Python, ces dernières apportent des informations sur la manière d'utiliser la fonction ainsi que sur son rôle.</p>

```python
def pair(n: int) -> bool:
    if n%2 == 0:
        return True
    else:    
        return False

print(pair(5))
print(pair(6))
print(pair(6.2))
```

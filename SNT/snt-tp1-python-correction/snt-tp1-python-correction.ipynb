## Partie 5 : Instructions conditionnelles
</div>

Les conditions Si/Sinon permettent d'executer un code diff√©rent selon si une condition est r√©alis√©e. On se donne l'algorithme suivant:

<img style="max-width:100%" src="https://capytale2.ac-paris.fr/web/sites/default/files/2022/01-22/18-01-33/Img4.png"/>

En Python, <code>Si condition alors</code> s'√©crira <code>if condition:</code> suivi d'un bloc indent√©. <code>Sinon</code> s'√©crit simplement <code>else:</code>.<br>
<b>Attention:</b> Pour tester l'√©galit√© en Python, il faut √©crire par exemple <code>if x==2:</code>: On met <b>deux</b> fois le symbole =.<br>
Voici le m√™me algorithme retranscrit en langage Python:

```python
def f(x):
    if x>=0:    #  >= signifie sup√©rieur ou √©gal
        y=x     #  Remarquons que cette instruction est indent√©e pour que Python reconnaisse ce qui doit √™tre execut√© dans le if
    else:
        y=-x
    return(y)   #  Cette instruction n'est pas indent√©e comme la ligne pr√©c√©dente: elle est execut√©e apr√®s les tests.

# Si x est positif, on le retourne, sinon on retourne son oppos√©. Ce programme supprime le - d'un nombre
```

```python
f(-5)
```

```python
f(5)
```

<h2 style="text-decoration:underline;" id="partie1">1- Boucle bornée (boucle <code>for</code>)</h2>

Le programme ci-dessous est un exemple simple d'utilisation de la boucle <code>for</code>.<br>
On répète ici 5 fois l'instruction <code>print("Bonjour")</code>.<br>
Le bloc d'instructions à répéter doit être **indenté**, c'est-à-dire décalé de quelques espaces vers la droite.<br>
La variable <code>i</code> est l'indice de boucle (ou variable de boucle ou compteur de boucle). On peut lui donner le nom que l'on veut. Son utilisation sera vue dans la partie <em><a href="#partie2">2- Utilisation du compteur de boucle</a></em>.

```python
for i in range(5):
    print("Bonjour")
```

<h2 style="text-decoration:underline;" id="partie2">2- Utilisation de la variable de boucle</h2>

Comme le montre l'exemple suivant, la variable de boucle (<code>i</code> dans l'exemple) est une variable ordinaire qui a initialement la valeur 0 et qui est incrémentée de 1 à chaque tour de la boucle. On peut choisir n'importe quel nombre pour cette variable. Des noms tels que <code>i, j, k, m, n, index</code> ou <code>compteur</code> sont fréquemment utilisés.<br>
Attention à la dernière valeur qui est 9 et non 10, car on a commencé à 0 ! Cependant, le nombre qui figure en paramètre de la fonction <code>range()</code> (ici 10) correspond bien au nombre de tours de boucle.<br>
La variable de boucle est aussi appelée "compteur de boucle". Elle permet de savoir combien de tours de boucle ont déjà été effectués.

```python
for i in range(10):
    print(i)
print(i)
```

<span style="text-decoration:underline;">Remarque</span> : 
- cet exemple montre aussi que la variable de boucle existe toujours après l'exécution de la boucle ;
- il est cependant déconseillé d'utiliser cette variable en dehors de la boucle ; <br>
- il est également déconseillé de modifier le contenu de la variable de boucle à l'intérieur de la boucle.

<strong>Les secrets de la fonction <code>range()</code></strong><br>

<details style="details[open] summary::-webkit-details-marker { color:#eaeaea; background:#eaeaea; outline: 2px solid #ADCA48;}"><summary style="summary::-webkit-details-marker { color:#ADCA48; background:#ADCA48;} ">Secret n°1</summary>
"range" signifie "intervalle" en anglais.<br>
<strong><code>range(n)</code> renvoie en réalité la liste de tous les entiers de 0 inclus à n exclu.</strong><br>    
Vous pouvez tester cela en tapant <code>list(range(8))</code>.<br>  
    "<code>for i in range(8) :</code>" signifie donc littéralement : <br>
    "pour i prenant successivement toutes les valeurs de 0 à 7, faire :"
</details>

<details style="details[open] summary::-webkit-details-marker { color:#eaeaea; background:#eaeaea; outline: 2px solid #ADCA48;}"><summary style="summary::-webkit-details-marker { color:#ADCA48; background:#ADCA48;} ">Secret n°2</summary>
    <strong>La fonction <code>range()</code> peut prendre deux paramètres en entrée.</strong><br>  
    Si <code>a</code> et <code>b</code> sont deux entiers, <code>range(a,b)</code> renvoie la liste de tous entiers compris entre <code>a</code> inclus et <code>b</code> exclu.<br>   
</details>

<details style="details[open] summary::-webkit-details-marker { color:#eaeaea; background:#eaeaea; outline: 2px solid #ADCA48;}"><summary style="summary::-webkit-details-marker { color:#ADCA48; background:#ADCA48;} ">Secret n°3</summary>
    <strong>La fonction <code>range()</code> peut même prendre trois paramètres en entrée.</strong><br>  
    Si <code>a</code>, <code>b</code> et <code>p</code> sont trois entiers, <code>range(a,b,c)</code> renvoie la liste de tous entiers compris entre <code>a</code> inclus et <code>b</code> exclu avec un pas <code>p</code>.<br>
    Si le pas <code>p</code> est négatif, les nombres de la liste seront données dans l'ordre décroissant.
</details>

<h2 style="text-decoration:underline;" id="partie3">3- Notion d'accumulateur</h2>

Le compteur de boucle permet de mettre en place un accumulateur, c'est-à-dire une variable qui est utilisée pour stocker les différents résultats successifs d'un calcul qui avance progressivement à chaque tour de boucle. On économise ainsi de la mémoire : on ne garde en mémoire par exemple que la somme des valeurs déjà rencontrées ou le dernier état, et non tous les états intermédiaires.<br>
La variable qui sert d'accumulateur doit être initialisée avant la boucle.

<strong>Exemple : calcul de la somme des entiers de 1 à 100</strong>

```python
somme = 0 # La variable somme joue le rôle de l'accumulateur
for i in range(101):
    somme = somme + i   # à chaque tour de boucle, on met à jour la valeur de l'accumulateur en lui ajoutant i
print(somme)      # On affiche la valeur finale de l'accumulateur
```

Remarque : il existe une formule qui permet de calculer cette somme sans effectuer toutes les additions, qui sera vue en cours de Spécialité Maths : $1+2+3+...+n = \frac{n(n+1)}{2}$.

<h2 style="text-decoration:underline;" id="partie4">4- Dessiner avec Turtle</h2>

La bibliothèque **turtle** permet de dessiner à l'écran.<br>
Le petit programme commenté ci-dessous permet d'obtenir un tracé intéressant :

```python
from turtle import *      # On importe la bibliothèque turtle

speed(10)                 # On règle la vitesse du tracé (un entier compris entre 1 et 10)
color('red', 'yellow')    # On fixe la couleur du tracé à 'rouge' et la couleur de rempissage à 'jaune'
begin_fill()
for i in range(36):
    forward(200)          # La tortue avance de 200 pixels 
    left(170)             # La tortue tourne vers la gauche de 170°
end_fill()
done()                    # Il faut terminer par done() pour lancer l'exécution du tracé    
```

<h2 style="text-decoration:underline;" id="partie1">1- Les conditions</h2>

**Structure conditionnelle de base : <em>si ... alors ... sinon ...</em>**

Cela correspond au pseudo-code suivant :
<code>
instructions A
Si condition vraie faire :
    instructions B
Sinon
    instructions C
instructions D
</code><br>
En Python, les mot-clés sont <code>if</code> (si) et <code>else</code> (sinon) :
<code>
A
if condition :
    B
else:
    C
D
</code>

Le programme ci-dessous adapte son comportement à la réponse fournie par l'utilisateur :

```python
âge = int(input("Veuillez entrer votre âge ! "))
if (âge>=18):
    print("Vous êtes majeur!")
else:
    print("Vous êtes mineur!")
print("Au revoir!")
```

**Structure conditionnelle incomplète : <em>si ... alors ... </em>**

La clause <code>else</code> est facultative. On obtient alors la structure suivante :<br>
<code>
    Bloc A
    Si condition vraie faire :
        Bloc B
    Bloc C
</code>
Ainsi, si la condition est vraie, on effectuera les blocs A-B-C.<br>
Si la condition n'est pas vraie, on effectuera les blocs A-C

Combinées avec une boucle, les conditions permettent de "filtrer" des valeurs, comme dans l'exemple ci-dessous qui compte le nombre de fois où un dé a donné 6 sur 1000 lancers :

```python
from random import *

compteur = 0
for i in range(1000):
    dé = randint(1,6)
    if dé == 6:
        compteur +=1
print("Le résultat 6 est sorti",compteur,"fois.")
```

Les booléens (ou variables booléennes) sont utilisés dans les conditions. En fait, l'expression qui suit le <code>if</code> doit toujours être une valeur ou une variable de type booléen :

```python
majeur = True    # False
if majeur == True:
    print("Vous êtes majeur.")
else:
    print("Vous êtes mineur")
```

<strong>Syntaxe raccourcie des conditions</strong><br>
Dans le programme précédent, <code>majeur == True</code> est en fait un booléen qui vaut <code>True</code> ou <code>False</code> selon la valeur de la variable <code>majeur</code>.<br>
Ainsi, les booléens permettent une syntaxe raccourcie et plus lisible dans les conditions : <br>
Au lieu d'écrire <code>if majeur == True:</code><br> on peut écrire : <code>if majeur:</code><br>
Il est préférable d'utiliser cette syntaxe raccourcie.

```python
majeur = True    # False
if majeur:
    print("Vous êtes majeur.")
else:
    print("Vous êtes mineur")
```

<h2 style="text-decoration:underline;" id="partie3">3- Les conditions multiples</h2>

On a parfois besoin de tester plusieurs conditions successivement comme dans l'exemple suivant.

**Le problème du pass sanitaire**<br>
On demande à une personne si elle est vaccinée et si elle a un test PCR négatif.<br>
Si elle est vaccinée, elle peut entrer.<br>
Sinon, elle ne peut entrer que si elle a un test PCR négatif.<br>

```python
vacciné = input("Etes-vous vacciné (O/N)?")=='O'
testNégatif = input("Avez-vous un test PCR négatif (O/N)?")=='O' 
          
if vacciné:
    print("Vous pouvez entrer...")
else:
    if testNégatif:
          print("Vous pouvez entrer!")
    else:
          print("Vous ne pouvez pas entrer!")
```

Le défaut de ce programme est qu'il est assez difficile à lire.<br>
En Python, on dispose d'une syntaxe plus lègère qui permet de clarifier un peu le programme : on peut condenser le <code>else if</code> en une seule ligne <code>elif</code>.<br>
Le programme ci-dessous fait exactement la même chose que le précédent mais nécessite une ligne de moins et fait moins appel à l'indentation dans les deux dernières lignes :

```python
vacciné = input("Etes-vous vacciné (O/N)?")=='O'
testNégatif = input("Avez-vous un test PCR négatif (O/N)?")=='O' 
          
if vacciné:
    print("Vous pouvez entrer...")
elif testNégatif:
    print("Vous pouvez entrer!")
else:
    print("Vous ne pouvez pas entrer!")
```

<div class="alert alert-warning" style="border-left: 3px solid"><strong>Exercice 46 : </strong> écrire un programme qui simule 1000 lancers de deux dés, et qui affiche le nombre de fois où une somme comprise entre 4 et 8 inclus a été obtenue.

</div>

```python
from random import *

compteur = 0
for i in range(1000):
    dé1 = randint(1,6)
    dé2 = randint(1,6)
    somme = dé1 + dé2
    if 4 <= somme and somme <= 8:
        compteur = compteur + 1
print("Une somme comprise entre 4 et 8 a été obtenue",compteur,"fois")
```

<span style="text-decoration:underline;">Remarque</span> : Python autorise ici une syntaxe plus courte que ne permettent pas les autres langages.
Pour tester si un nombre est compris entre 3 et 7, on peut écrire <code>3 &lt; x &lt; 7</code> au lieu de <code>3 &lt; x and x &lt; 7</code>.

**Nouvelle résolution du problème du pass sanitaire**<br>
Avec les booléens, on peut résoudre le problème du pass sanitaire de façon plus élégante, plus claire et plus courte qu'avec des condtions imbriquées :

```python
vacciné = input("Etes-vous vacciné (O/N)?")=='O'
testNégatif = input("Avez-vous un test PCR négatif (O/N)?")=='O' 
          
if vacciné or testNégatif:
    print("Vous pouvez entrer!")
else:    
    print("Vous ne pouvez pas entrer!")
```
